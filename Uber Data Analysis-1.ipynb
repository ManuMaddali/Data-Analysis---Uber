## Libraries: Pandas, Seaborn

```python
%pylab inline
import pandas
import seaborn
```

```python
data = pandas.read_csv('Desktop/uber-raw-data-apr14.csv')
```

```python
data.tail()
```

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.tail()
```

```python
def get_dom(dt):
    return dt.day
data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()
data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour
data['hour'] = data["Date/Time"].map(get_hour)

data.tail()

    
```

# analysis

# analyze DoM

```python
hist(data.dom, bins = 31, rwidth = .8, range=(0.5, 30.5), color = 'green')
xlabel('date of the month ')
ylabel('amount/people')
title('Frequency DoM - Uber - Apr-2014')
```

```python
#for k, rows in data.groupby('dom'):
#   print((k, len(rows)))
def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
bar(range(1, 31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1, 31), by_date_sorted, color = 'yellow')
xticks(range(1, 31), by_date_sorted.index)
xlabel('date of the month ')
ylabel('amount/people')
title('Frequency DoM - Uber - Apr-2014')
;
```
